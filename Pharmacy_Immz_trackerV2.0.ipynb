```python
total_patients = int(input("Enter the number of patients: "))  # Fixed: convert to int early

def is_eligible(age):
    return age >= 18

def can_vaccinate(has_appointment):
    return has_appointment.lower().strip() == 'yes'

patients_list = []
eligible_count = 0
ineligible_count = 0
total_vaccines_given = 0

for i in range(total_patients):
    print(f"\nEntering info for patient {i + 1}:")
    
    name = input("Name: ")
    
    age_input = input("Age: ")
    while not age_input.isdigit():
        print("Please enter a valid number.")
        age_input = input("Age: ")
    age = int(age_input)
    
    has_appt_input = input("Has appointment? (yes/no): ")
    while has_appt_input.lower().strip() not in ['yes', 'no']:
        print("Please enter 'yes' or 'no'.")
        has_appt_input = input("Has appointment? (yes/no): ")
    
    has_appointment = has_appt_input.lower().strip() == 'yes'

    if has_appointment:
        eligible_count += 1
    else:
        ineligible_count += 1

    vaccine_list = []
    vaccine_count = int(input("Number of vaccines today: "))
    total_vaccines_given += vaccine_count

    for v in range(vaccine_count):
        v_name = input(f"  Vaccine {v+1} name: ")
        v_date = input(f"  Vaccine {v+1} date (YYYY-MM-DD): ")
        vaccine_list.append({'name': v_name, 'date': v_date})

    patient = {
        'name': name,
        'age': age,
        'has_appointment': has_appointment,
        'is_eligible': is_eligible(age),
        'can_vaccinate': can_vaccinate(has_appt_input),
        'vaccine_count': vaccine_count,
        'vaccines': vaccine_list
    }
    patients_list.append(patient)

# === Reporting ===
print("\nðŸ“‹ Patient Summary Report:")
for patient in patients_list:
    print(f"Name: {patient['name']}")
    print(f"Age: {patient['age']}")
    print(f"Has Appointment: {'Yes' if patient['has_appointment'] else 'No'}")
    print(f"Eligible: {'Yes' if patient['is_eligible'] else 'No'}")
    print(f"Can Vaccinate: {'Yes' if patient['can_vaccinate'] else 'No'}")
    print(f"Vaccines Given: {patient['vaccine_count']}")
    for v in patient['vaccines']:
        print(f"  - {v['name']} on {v['date']}")
    print("â€”" * 30)

# === Totals ===
print("\nðŸ“Š Final Stats:")
print(f"Total Patients: {len(patients_list)}")
print(f"Total Vaccines Administered: {total_vaccines_given}")
print(f"Eligible: {eligible_count} | Ineligible: {ineligible_count}")
print(f"Eligibility Rate: {round((eligible_count / total_patients) * 100, 1)}%")
print("âœ… Thank you for using the pharmacy immunization system!")
```
