```python
total_patients= input("Enter the number of patients: ") # gathering number of patients to create the loop

def is_eligible(age):           #defining eligiblity functions
    return age >= 18
def can_vaccinate(has_appointment):
    return has_appointment.lower().strip() == 'yes'

patients_list = [] # list to store patient data

eligible_count = 0      #counters for eligibility
ineligible_count = 0

for i in range(int(total_patients)):  # loop to gather patient data
    vaccine_list = []  # list to store vaccine data
    name = input(f"Enter the name of patient {i + 1}: ")
    age_input= input(f"what is the patient's age?")
    while not age_input.isdigit():
        print("Please enter a valid age (a number).")
        age_input = input(f"Enter the age of patient {i + 1}: ")
    age = int(age_input)
    has_apointment_input = input(f"Does the patient have an appointment? (yes/no): ")
    while has_apointment_input.lower().strip() not in ['yes', 'no']:
        print("Please enter 'yes' or 'no'.")
        has_apointment_input = input(f"Does the patient have an appointment? (yes/no): ")    
    has_appointment = has_apointment_input.lower().strip() == 'yes'
    if has_appointment:
        eligible_count += 1
    else:
        ineligible_count += 1

    vaccine_count = int(input(f"How many vaccines is {name} getting today? "))
    for v in range(int(vaccine_count)):
        vaccine_name = input(f"Enter the name of vaccine {v + 1}  ")
        vaccine_date = input(f"Enter the date of vaccine {v + 1} (YYYY-MM-DD): ")
        while not vaccine_date.isdigit():
            print("Please enter a valid date.")
            vaccine_date = input(f"Enter the date of vaccine {v + 1} (YYYY-MM-DD): ")

        vaccine={ 
            'name' : vaccine_name,
            'date': vaccine_date

        }
        vaccine_list.append(vaccine)

    patient = {
        'name': name,
        'age': age,
        'has_appointment': has_appointment,
        'vaccine_count': vaccine_count,
        'is_eligible': is_eligible(age),
        'can_vaccinate': can_vaccinate(has_apointment_input),
        'vaccines': vaccine_list.copy()  # copy the vaccine list for each patient
    }
    patients_list.append(patient)


# Displaying the patient data
print("\nPatient Data:")
for patient in patients_list:
    print(f"Name: {patient['name']}")
    print(f"Age: {patient['age']}")
    print(f"Has Appointment: {'Yes' if patient['has_appointment'] else 'No'}")
    print(f"Eligible for Vaccination: {'Yes' if patient['is_eligible'] else 'No'}")
    print(f"Can Vaccinate Today: {'Yes' if patient['can_vaccinate'] else 'No'}")
    print(f"Number of Vaccines: {patient['vaccine_count']}")
    print("Vaccines:")
    for vaccine in patient['vaccines']:
        print(f"  - {vaccine['name']} on {vaccine['date']}")
    print()  # Print a newline for better readability

print("Total number of patients:", len(patients_list))
print("Total number of vaccines administered:",(total_patients))
print("Eligible patients:", eligible_count)
print("Ineligible patients:", ineligible_count)
if len(patients_list) > 0:
    print("Eligibility percentage:", round((eligible_count / len(patients_list)) * 100, 1), "%")
else:
    print("No patients entered. Cannot calculate eligibility percentage.")
print("Thank you for using the vaccination system, and have a great day!")
    
```
