```python
total_patients = int(input("Enter the number of patients vaccinated today: "))


def is_eligible(age):
    return age >= 18

def can_vaccinate(has_appointment: bool):
    return has_appointment


patients_list = []

eligible_count = 0
ineligible_count = 0
total_vaccines = 0

for i in range(total_patients):
    name = input(f"Enter the name of patient {i + 1}: ")
    age_input = input(f"Enter the age of patient {i + 1}: ")
    while not age_input.isdigit() or int(age_input) < 0:
        print("Invalid input. Please enter a valid age.")
        age_input = input(f"Enter the age of patient {i + 1}: ")
    age = int(age_input)
    has_appointment_input = input(f"Does patient {i + 1} have an appointment? (yes/no): ")
    while has_appointment_input.lower().strip() not in ['yes', 'no']:
        print("Invalid input. Please enter 'yes' or 'no'.")
        has_appointment_input = input(f"Does patient {i + 1} have an appointment? (yes/no): ")
    has_appointment = has_appointment_input.lower().strip() == 'yes'
    if has_appointment and is_eligible(age):
        eligible_count += 1
    else:
        ineligible_count += 1

    vaccine_list = []
    vaccine_count_input = input("How many doses of vaccine has the patient received? ")
    while not vaccine_count_input.isdigit() or int(vaccine_count_input) < 0:
        print("Invalid input. Enter a valid non-negative number.")
        vaccine_count_input = input("How many doses of vaccine has the patient received? ")
    vaccine_count = int(vaccine_count_input)
    for v in range(vaccine_count):
        vaccine_name = input(f"Enter the name of vaccine dose {v + 1} for patient {i + 1}: ")
        vaccine_date = input(f"Enter the date of vaccine dose {v + 1} for patient {i + 1} (YYYY-MM-DD): ")
        while len(vaccine_date) != 10 or vaccine_date[4] != '-' or vaccine_date[7] != '-':
            print("Invalid date format. Please enter the date in YYYY-MM-DD format.")
            vaccine_date = input(f"Enter the date of vaccine dose {v + 1} for patient {i + 1} (YYYY-MM-DD): ")
        vaccine = {
            'name': vaccine_name,
            'date': vaccine_date
        }
        vaccine_list.append(vaccine)

    patient = {
        'name': name,
        'age': age,
        'has_appointment': has_appointment,
        'is_eligible': is_eligible(age),
        "can_vaccinate": has_appointment,
        'vaccine_count': vaccine_count,
        'vaccines': vaccine_list
    }

    patients_list.append(patient)
    with open('patients.txt', "a") as file:
        file.write(f"Patient Name: {name}\n")
        file.write(f"Age: {age}\n")
        file.write(f"Has Appointment: {'Yes' if has_appointment else 'No'}\n")
        file.write(f"Eligible: {'Yes' if is_eligible(age) else 'No'}\n")
        file.write(f"Vaccines:\n")
        for v in vaccine_list:
            file.write(f"  - {v['name']} on {v['date']}\n")
        file.write("\n")

    if is_eligible(age) and can_vaccinate(has_appointment) and vaccine_count > 0:
        total_vaccines += patient['vaccine_count']
    
    want_toread_input = input(f"Does patient {i + 1} want to read the vaccine log? (yes/no): ")
    while want_toread_input.lower().strip() not in ['yes', 'no']:
        print("Invalid input. Please enter 'yes' or 'no'.")
        want_toread_input = input(f"Does patient {i + 1} want to read the vaccine log? (yes/no): ")
    want_toread = want_toread_input.lower().strip() == 'yes'

    if want_toread:
       print("\n--- Vaccine Log ---")
       with open('patients.txt', "r") as file:
            print(file.read())







print("Vaccination Summary:")
for patient in patients_list:
    print(f"Patient Name: {patient['name']}")
    print(f"Age: {patient['age']}")
    print(f"Eligible: {patient['is_eligible']}")
    print(f"Has Appointment: {patient['has_appointment']}")
    print(f"Vaccine Count: {patient['vaccine_count']}")
    print(f"Vaccine details: {patient['vaccines']}")

print(f"Total eligible patients: {eligible_count}")
print(f"Total ineligible patients: {ineligible_count}")
print(f"Total vaccines administered: {total_vaccines}")

print("Vaccination log has been saved to 'patients.txt'.")
print("Thank you for using the vaccination system!")
```
