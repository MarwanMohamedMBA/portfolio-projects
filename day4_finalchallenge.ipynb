```python
#day 4 final challenge. Pharmacy Immunization Tracker v1.0


#asking for number of patients

total_patients= int(input("how many patients vaccinated today: "))

#eligibility functions
def is_eligible(age):
    return age >= 18

def can_vaccinate(has_appointment, vaccine_instock):
    return has_appointment and vaccine_instock


#creating empty patint list

patients_list = []


#collecting patient info

for i in range(total_patients):
    print(f"\n Entering info for patient {i+1}")

    name = input("What is your name? ")

    #validating patient age

    age_input = input("How old are you? ")
    while not age_input.isdigit():
        print("Please enter a valid number.")
        age_input = input("How old are you? ")
    age = int(age_input)

    has_appointment_input = input("Do you have an appointment? (yes/no): ").lower()
    has_appointment = has_appointment_input == "yes"
    vaccine_type = input("Which vaccine are you here for today? ")
    vaccine_instock = input("Is the vaccine in stock? (yes/no) ").lower() == "yes"


        # Store patient info in a dictionary
    patient = {
        "name": name,
        "age": age,
        "has_appointment": has_appointment,
        "vaccine_type": vaccine_type,
        "vaccine_instock" : vaccine_instock

    }
        # Add the dictionary to the list
    patients_list.append(patient)



        
#printing summary reports

print("\n📋 Patient Summary Report:")
total_vaccinated = 0

for patient in patients_list:
    name = patient["name"]
    age = patient["age"]
    has_appointment = patient["has_appointment"]
    vaccine_type = patient["vaccine_type"]
    vaccine_instock = patient["vaccine_instock"]

    if is_eligible(age) and can_vaccinate(has_appointment, vaccine_instock):
        print(f"✅ {name} ({age}) is ELIGIBLE and received the {vaccine_type} vaccine.")
        total_vaccinated += 1
    else:
        print(f"❌ {name} ({age}) is NOT eligible or could not be vaccinated.")

print(f"\n💉 Total vaccines successfully administered: {total_vaccinated}")
```
